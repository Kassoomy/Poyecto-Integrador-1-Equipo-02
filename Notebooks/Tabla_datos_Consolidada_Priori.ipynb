```
def tryDatetime(time2Check):
    try: 
        return (datetime.min + timedelta(minutes=(((time2Check-datetime(1900, 1, 1, 0, 0)).total_seconds()+(24*3600))/60/60))).time()
    except (NameError, TypeError, ValueError, AttributeError) as error: 
        try:
            return time(hour=0, minute=time2Check.hour, second=time2Check.minute, microsecond=time2Check.second) 
        except (NameError, TypeError, ValueError, AttributeError) as error: 
            return time2Check
```

```
dfStandard = df_datos_priori.copy()
columns2Change = [  'Tiempo_Recomendación1',
       'Tiempo_Recomendación2', 'Tiempo_Recomendación3', 'Tiempo_Publicidad_1',
       'Tiempo_Publicidad_2', 'Tiempo_Publicidad_3', 'Tiempo_Publicidad_4',
       'Tiempo_Publicidad_5', 'Tiempo_Publicidad_6', 'Tiempo_Publicidad_7',
       'Tiempo_Publicidad_8']
for currColumn in columns2Change:
    dfStandard[currColumn] = dfStandard[currColumn].apply(lambda cell:  tryDatetime(cell))
dfStandard = dfStandard.reset_index(drop=True)
dfStandard
```

```
columnsPublicidad = ['Tiempo_Publicidad_1',
       'Tiempo_Publicidad_2', 'Tiempo_Publicidad_3', 'Tiempo_Publicidad_4',
       'Tiempo_Publicidad_5', 'Tiempo_Publicidad_6', 'Tiempo_Publicidad_7',
       'Tiempo_Publicidad_8']

columnsRecomendacion = ['Tiempo_Recomendación1',
       'Tiempo_Recomendación2', 'Tiempo_Recomendación3']

dfStandard['Cantida Publicidad'] = 0
dfStandard['Cantida Recomendacion'] = 0
dfStandard['Duracion Minutos'] = 0.00
dfStandard['Densidad_Publicitaria'] = 0.00
dfStandard['Densidad_Recomendacion'] = 0.00

for rows in range(len(dfStandard['likes'])):
    dfStandard.at[rows,'Cantida Publicidad'] = int(8-dfStandard.iloc[rows][columnsPublicidad].isnull().sum())
    dfStandard.at[rows,'Cantida Recomendacion'] = int(3-dfStandard.iloc[rows][columnsRecomendacion].isnull().sum())
    dfStandard.at[rows, 'Duracion Minutos'] = (dfStandard.iloc[rows]['duracion_video_minutos'] + dfStandard.iloc[rows]['duracion_video_segundos']/60)
    dfStandard.at[rows, 'Densidad_Publicitaria'] = round(((dfStandard.iloc[rows]['Duracion Minutos']/dfStandard.iloc[rows]['Cantida Publicidad'])**-1),3)
    dfStandard.at[rows, 'Densidad_Recomendacion'] = round(((dfStandard.iloc[rows]['Duracion Minutos']/dfStandard.iloc[rows]['Cantida Recomendacion'])**-1),3)

dfStandard
```

```
for nro in range(1,4):
    dfStandard['minutos_recom'+str(nro)] = dfStandard['Tiempo_Recomendación'+str(nro)].astype(str).str[3:5]
    dfStandard['minutos_recom'+str(nro)] = pd.to_numeric(dfStandard['minutos_recom'+str(nro)], errors='coerce')
    dfStandard['seg_recom'+str(nro)] = dfStandard['Tiempo_Recomendación'+str(nro)].astype(str).str[6:8]
    dfStandard['seg_recom'+str(nro)] = pd.to_numeric(dfStandard['seg_recom'+str(nro)], errors='coerce')
    dfStandard['minutos_recom'+str(nro)] = dfStandard['minutos_recom'+str(nro)] + (dfStandard['seg_recom'+str(nro)]/60)

for nro in range(1,9):
    dfStandard['minutos_publi'+str(nro)] = dfStandard['Tiempo_Publicidad_'+str(nro)].astype(str).str[3:5]
    dfStandard['minutos_publi'+str(nro)] = pd.to_numeric(dfStandard['minutos_publi'+str(nro)], errors='coerce')
    dfStandard['seg_publi'+str(nro)] = dfStandard['Tiempo_Publicidad_'+str(nro)].astype(str).str[6:8]
    dfStandard['seg_publi'+str(nro)] = pd.to_numeric(dfStandard['seg_publi'+str(nro)], errors='coerce')
    dfStandard['minutos_publi'+str(nro)] = dfStandard['minutos_publi'+str(nro)] + (dfStandard['seg_publi'+str(nro)]/60)
```

```
for nro in range(1,4):
    dfStandard['porc_min_recom'+str(nro)] = dfStandard['minutos_recom'+str(nro)] / dfStandard['Duracion Minutos']

for nro in range(1,9):
    dfStandard['porc_min_publi'+str(nro)] = dfStandard['minutos_publi'+str(nro)] / dfStandard['Duracion Minutos']
```

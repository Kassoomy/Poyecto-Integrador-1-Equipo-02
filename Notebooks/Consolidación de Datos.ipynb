```
# Librerias

!pip install openpyxl
import openpyxl
import pandas as pd
from glob import glob
from zipfile import ZipFile
import seaborn as sns
import matplotlib.pyplot as plt


# Quitar tildes de columnas Dataframe
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s

# Estilo plots
sns.set()
```

## Consolidación de Datos

```
# Tablas de interés:

files_zip = ['Datos de la tabla.csv', 'Datos del gráfico.csv'] # 'Totales.csv'

# Carpetas principales:

carpetas_videos = glob('/work/datasets/datasets/*')
print(len(carpetas_videos))  # debe coincidir con el número de videos en la carpeta

archivos_iterar = []  # todas las rutas quedan en esta carpeta

for posicion_carpeta_principal in range(len(carpetas_videos)):
    
    for i in glob(carpetas_videos[posicion_carpeta_principal]+'/*.zip'):
        archivos_iterar.append(i)

len(archivos_iterar)
```

### Datos de la tabla

```
# Datos de la tabla

data_tabla_estado_suscripcion = [] 
data_tabla_tipo_dispositivo = []
data_tabla_ciudades = []
data_tabla_area_geogr = [] 
data_tabla_edad_espect = []
data_tabla_contenido = []
data_tabla_fecha = []
data_tabla_fuente_trafico = [] 
data_tabla_sexo_espect = []

listas_no_nulas_posi = []

for tabla in archivos_iterar:

    with ZipFile(tabla) as myzip:

        if 'Datos de la tabla.csv' in myzip.namelist():
            data = myzip.open('Datos de la tabla.csv')

            df = pd.read_csv(data)
            df.columns = df.columns.str.lower()
            df.columns = df.columns.str.replace(' ', '_')
            df.columns = [normalize(i) for i in list(df.columns)]

            # creando columnas 
            file = tabla

            df['fecha_publicacion'] = file.split('/')[4][:8]
            df['nombre_video'] = file.split('/')[4][9:]


            if 'Estado de suscrip' in tabla:
                data_tabla_estado_suscripcion.append(df)

            elif 'Tipo de dispositivo' in tabla:
                data_tabla_tipo_dispositivo.append(df)
            
            elif 'Ciudades' in tabla:
                data_tabla_ciudades.append(df)

            elif 'geogr' in tabla:
                data_tabla_area_geogr.append(df)
            
            elif 'Edad' in tabla:
                data_tabla_edad_espect.append(df)

            elif 'Contenido' in tabla:
                data_tabla_contenido.append(df)
            
            elif 'Fecha' in tabla:
                data_tabla_fecha.append(df)

            elif 'Fuente de tr' in tabla:
                data_tabla_fuente_trafico.append(df)
            
            elif 'Sexo' in tabla:
                data_tabla_sexo_espect.append(df)

# Exportando datos:

data_tabla_listas = [data_tabla_estado_suscripcion,
                    data_tabla_tipo_dispositivo,
                    data_tabla_ciudades,
                    data_tabla_area_geogr, 
                    data_tabla_edad_espect, 
                    data_tabla_contenido, 
                    data_tabla_fecha, 
                    data_tabla_fuente_trafico,
                    data_tabla_sexo_espect]

# Si hay listas vacías porque no tienen datos eliminarlas:

for i in range(len(data_tabla_listas)):
    if len(data_tabla_listas[i]) > 0:
        
        listas_no_nulas_posi.append(i)

data_tabla_listas = [data_tabla_listas[i] for i in listas_no_nulas_posi]


data_tabla_names = ['data_tabla_estado_suscripcion',
                    'data_tabla_tipo_dispositivo',
                    'data_tabla_ciudades',
                    'data_tabla_area_geogr', 
                    'data_tabla_edad_espect', 
                    'data_tabla_contenido', 
                    'data_tabla_fecha', 
                    'data_tabla_fuente_trafico',
                    'data_tabla_sexo_espect']


data_tabla_names = [data_tabla_names[i] for i in listas_no_nulas_posi]

rutas_data_tabla = ['data_consolidada/datos_tabla/df_' + str(i) + '.csv' for i in data_tabla_names]    

for lista_data_tabla, ruta_data_tabla in zip(data_tabla_listas, rutas_data_tabla):
    pd.concat(lista_data_tabla).to_csv(ruta_data_tabla, index=False,  encoding='utf-8-sig')
```
